# Plotting HowTo: ( Matplotlib ) 

**Prerequisites** : Variables, lists, running code ...( Modules ) 

**Matplotlib** is a python library used to create graphs and charts for visualization. 
It creates professional-looking visualizations used while working on Science projects, math assignments or
just curious about data.

- Matplotlib is a python module for creating graphs and chats in 2D or 3D plots. The module comes with many
  functions that help creating many types of graphs and layouts. 

- It's like digital piece of paper, You provide data to its co-ordinates and it handles the drawing. 

- Common charts:
    - Line graphs 
    - Bar graphs 
    - Pie Charts 
    - Scatter plots

## Getting Started:

- Import the library module: and call it 'plt' or any other short name 

```python 
import matplotlib.pyplot as plt 
```

- The `matplotlib` library contains many modules 
    - `matplotlib.pyplot` : most used and important one 
    - `matplotlib.figure` : Controls the whole figure ( window ) ( controls individual plots inside figure )
    - `matplotlib.colors` : Handle colors and maps 
    - `matplotlib.cm` : colormaps ( used in heatmaps, images )
    - `matplotlib.style` : pre-made graph styles 
    - `matplotlib.animation` : create animation. ( advanced )
    - `matplotlib.ticker` : control tick marks on axes ( advanced )

## How to use the library for plotting: 

- Step 1: import the plotting module from matplotlib

- Step 2: Collect or prepare the data for x,y,z   (ex: as lists, arrays,  tuples, sets ..)

- Step 3: Plot the data using plt.plot() 

````python
# List Graph 
import matplotlib.pyplot as plt
scores = [65, 70, 78, 85, 90]
tests = [1, 2, 3, 4, 5]

plt.plot(tests, scores, marker='o')
plt.xlabel("Test Number")
plt.ylabel("Score")
plt.title("Math Test Scores Over Time")
plt.grid()

plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Create some data (x-values from 0 to 10, 100 points)
x = np.linspace(0, 10, 100)  # 100 evenly spaced numbers from 0 to 10
y = x ** 2  # y = x (a parabola!)

# Create the plot
plt.figure(figsize=(8, 6))  # Set figure size (width, height in inches)
plt.plot(x, y, 'b-', linewidth=2, label='y = x') #, marker="^")  # 'b-' means blue line
plt.plot(x, x, 'r--', label='y = x')  # 'r--' means red dashed line

# Add labels and title
plt.xlabel('X values', fontsize=12)
plt.ylabel('Y values', fontsize=12)
plt.title('Our First Matplotlib Graph!', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)  # Add grid with transparency
plt.legend()  # Show the legend
plt.show()  # Display the plot
````

````python
# when data points have error corrections ( or margins ) using indicator to show
import matplotlib.pyplot as plt
import numpy as np

# Sample data: Plant growth with fertilizer
days = np.array([1, 2, 3, 4, 5, 6, 7])
height = np.array([5.1, 7.2, 9.8, 12.5, 14.9, 17.3, 19.8])  # in cm

# Error values (uncertainty in measurement)
error = np.array([0.5, 0.6, 0.7, 0.8, 0.9, 1.0, 1.1])  #  this much

plt.figure(figsize=(9, 6))
plt.errorbar(days, height, yerr=error, 
             fmt='o',           # 'o' for circles at data points
             color='darkgreen',
             ecolor='red',      # Error bar color
             elinewidth=2,      # Error bar line width
             capsize=5,         # Size of the "caps" on the I
             capthick=2,        # Thickness of caps
             markersize=8,
             label='Plant Height  Error')
plt.plot(days, height, 'r')

plt.xlabel('Days After Planting', fontsize=12)
plt.ylabel('Height (cm)', fontsize=12)
plt.title('Plant Growth Experiment with Error Bars', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
````

````python
import matplotlib.pyplot as plt
import numpy as np

# Favorite subjects survey
subjects = ['Math', 'Science', 'English', 'History', 'Art']
votes = [15, 12, 8, 10, 7]
colors = ['#FF6B6B', '#4ECDC4', '#FFD166', '#06D6A0', '#118AB2']

# Calculate realistic error based on sample size
# For survey data, margin of error is often calculated as:
# Error = 1.96 * sqrt(p*(1-p)/n) * total_votes
# Where p = proportion, n = sample size

total_votes = sum(votes)
errors = []

for vote in votes:
    proportion = vote / total_votes
    n = 52  # Total number of students surveyed
    # Standard error for proportion
    std_error = np.sqrt(proportion * (1 - proportion) / n)
    # Margin of error at 95% confidence
    margin_error = 1.96 * std_error * total_votes
    errors.append(round(margin_error, 1))

print(f"Calculated errors: {errors}")

plt.figure(figsize=(10, 7))

# Create bar chart with error bars
bars = plt.bar(subjects, votes, 
               color=colors, 
               edgecolor='black',
               linewidth=1.5,
               yerr=errors,
               capsize=12,
               error_kw={'elinewidth': 2.5,
                         'ecolor': '#333333',
                         'capthick': 2})

# Add value labels with errors
for bar, error in zip(bars, errors):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + error + 0.6,
             f'{int(height)} ({error})', 
             ha='center', 
             fontweight='bold',
             fontsize=10,
             color='black')

# Customize the plot
plt.xlabel('Subjects', fontsize=13)
plt.ylabel('Number of Votes', fontsize=13)
plt.title('Favorite Subjects Survey (n=52 students)\nwith 95% Confidence Interval', 
          fontsize=15, fontweight='bold', pad=15)
plt.ylim(0, max(votes) + max(errors) + 2)
plt.grid(True, alpha=0.2, axis='y', linestyle='--')

# Add legend for error bars
plt.errorbar([], [], yerr=[1], fmt='none',
             ecolor='black', elinewidth=2, capsize=8,
             label='Margin of Error (95% CI)')
plt.legend(loc='upper right', framealpha=0.9)

plt.tight_layout()
plt.show()
````
